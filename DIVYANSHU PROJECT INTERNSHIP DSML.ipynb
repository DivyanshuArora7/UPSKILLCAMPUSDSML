```python
# Bar Chart
plt.title('Yield (Quintal/ Hectare)')
plt.bar(df1['Crop'], df1['Yield (Quintal/ Hectare) '])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.style.use('fast')
fig,axs = plt.subplots(figsize=(10,6))
crop_wise_yield = df1.groupby(['Crop']).sum()['Yield (Quintal/ Hectare) ']
plt.plot(crop_wise_yield)
crop_wise_production = df1.groupby(['Crop']).sum()['Cost of Production (`/Quintal) C2']/10
plt.plot(crop_wise_production)
plt.ylabel('Count')
plt.xlabel('Crop', fontsize=15)
plt.xticks(rotation ='vertical')
plt.legend()
plt.show()
```

```python
# bar chart
plt.title('states with highest crop production')
a=df1['State'].value_counts()
plt.barh(a.index, a.values)
plt.show()
```

```python
plt.title('distribution of crops')
b=df3.Crop.value_counts()
plt.barh(b.index, b.values)
plt.show()
```

```python
# countplot
columns=df1.columns
df1.groupby('Crop')[columns[:-1]].sum().plot(kind='bar', figsize=(8,6))
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you want to use all columns
cols = df1.columns  

# Group by 'State' and sum the values in the last column
state_wise_sum = df1.groupby('State')[cols[-1]].sum()

# Plot the pie chart
plt.figure(figsize=(5, 5))
plt.pie(state_wise_sum, labels=state_wise_sum.index, autopct='%1.1f%%', startangle=140)
plt.title('State-wise ' + cols[-1], color='red', fontsize=12)

# Display the pie chart
plt.show()
```
